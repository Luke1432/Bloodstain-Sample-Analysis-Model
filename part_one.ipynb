```python
import skimage as ski
import matplotlib.pyplot as plt
import os
import pandas as pd

blunt_folder_path='resized\\SIZE_120_rescaled_max_area_1024\\120_blunt'
gun_folder_path='resized\\SIZE_120_rescaled_max_area_1024\\120_gun'

gun_images=os.listdir(gun_folder_path)
blunt_images=os.listdir(blunt_folder_path)

print('Number of images in the gun folder:', len(gun_images))
print('Number of images in the blunt folder:', len(blunt_images))

for value in [blunt_folder_path, gun_folder_path]:
        i=0
        for filename in os.listdir(value):
            if i==3:
                 break
            if filename.endswith(".jpg"):
                image = ski.io.imread(os.path.join(value, filename))
                plt.imshow(image, cmap='gray')
                plt.title(filename)
                plt.show()
            else:
                continue
            i+=1
```
